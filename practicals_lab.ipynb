Question 1.
Define a function, which takes in two strings A and B. Reverses the first string A, adds it to B, and returns the final string.

```
def revandconc(A, B):
    # Reverse the string A
    reversed_A = A[::-1]

    # Concatenate the reversed A with B
    result = reversed_A + B

    return result
A = "Hello"
B = "World"
result = revandconc(A, B)
print(result)

```

Question2.
Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]
Answer : the female worker that worked the most hours is Penny and she should be paid 18400

```
def calculate_highest_paid_female_worker(work_hours):
    max_hours = 0
    highest_paid_female = None

    for name, hours, gender in work_hours:
        if gender == 'F' and hours > max_hours:
            max_hours = hours
            highest_paid_female = name

    if highest_paid_female:
        pay = max_hours * 20
        return highest_paid_female, pay
    else:
        return "No female workers found."

# Given work_hours list
work_hours = [('Abby', 100, 'F'), ('Billy', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'),
              ('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]

# Calculate and print the result
result = calculate_highest_paid_female_worker(work_hours)
if isinstance(result, tuple):
    name, pay = result
    print(f"The female worker that worked the most hours is {name} and she should be paid ${pay}.")
else:
    print(result)
```
